```python
# Modell erstellen und l√∂sen
from gurobipy import *
m = Model()
x0 = m.addVar()
x1 = m.addVar()
m.update()
m.addConstr(x1 <= 4 - x0)
m.addConstr(x1 <= x0)
m.addConstr(x1 >= 0.5 + 0.5 * x0 )
m.setObjective(x1, GRB.MINIMIZE)
m.optimize()
```

```python
# Zul√§ssiger Bereich und optimale L√∂sung grafisch darstellen
import matplotlib.pyplot as pyplot
pyplot.plot([0,4], [0,4])
pyplot.plot([4,0], [0,4])
pyplot.plot([0,4], [0.5,2.5])
pyplot.plot([x0.x], [x1.x], 'ro')
pyplot.show()
```
